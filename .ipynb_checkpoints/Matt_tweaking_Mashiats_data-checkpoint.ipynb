```python
#load all packages
import GEOparse
import pandas as pd
from time import time
import numpy as np
import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import seaborn as sns
from sklearn import preprocessing

def bench_k_means(estimator, name, data):
    t0 = time()
    estimator.fit(data)
    print('%-9s\t%.2fs\t%i\t%.3f\t%.3f\t%.3f\t%.3f\t%.3f\t%.3f'
          % (name, (time() - t0), estimator.inertia_,
             metrics.homogeneity_score(labels, estimator.labels_),
             metrics.completeness_score(labels, estimator.labels_),
             metrics.v_measure_score(labels, estimator.labels_),
             metrics.adjusted_rand_score(labels, estimator.labels_),
             metrics.adjusted_mutual_info_score(labels,  estimator.labels_,
                                                average_method='arithmetic'),
             metrics.silhouette_score(data, estimator.labels_,
                                      metric='euclidean',
                                      sample_size=sample_size)))

#load the data
gse = GEOparse.get_GEO(geo="GSE10245", destdir="./")
gse_table = gse.pivot_samples('VALUE')
normalized_data = preprocessing.normalize(gse_table, norm='l2')
#normalized_data
scale_data = scale(normalized_data)
scale_data = scale_data.transpose()
n_samples, n_features = scale_data.shape
n_digits = 2


labels = gse_table.columns.values
sample_size = 58

bench_k_means(KMeans(init='k-means++', n_clusters=n_digits, n_init=10),
              name="k-means++", data=scale_data)

bench_k_means(KMeans(init='random', n_clusters=n_digits, n_init=10),
              name="random", data=scale_data)

pca = PCA(n_components=n_digits).fit(scale_data)
bench_k_means(KMeans(init=pca.components_, n_clusters=n_digits, n_init=1),
              name="PCA-based",
              data=scale_data)
print(82 * '_')


reduced_data = PCA(n_components=2).fit_transform(scale_data)
kmeans = KMeans(init='k-means++', n_clusters=n_digits, random_state=0, n_init=10)
kmeans.fit(reduced_data)


# Step size of the mesh. Decrease to increase the quality of the VQ.
h = .02     # point in the mesh [x_min, x_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired,
           aspect='auto', origin='lower')

plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=10)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=169, linewidths=3,
            color='w', zorder=10)
plt.title('K-means clustering on the digits dataset (PCA-reduced data)\n'
          'Centroids are marked with white cross')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
    
print(kmeans.fit(reduced_data).labels_.tolist())
print(gse_table.columns.tolist())

pheno_data=gse.phenotype_data[["title", "characteristics_ch1.0.disease state"]]

pheno_data['ClustersfromPCA'] = kmeans.fit(reduced_data).labels_.tolist()

pheno_data
```

```python
def cluster_output(kmeans_dataframe, cancer_one, cancer_two, cancer_type, cluster_number):
    #iteration through dataframe build from answer found on 
    #https://stackoverflow.com/questions/16476924/how-to-iterate-over-rows-in-a-dataframe-in-pandas?answertab=oldest#tab-top 
    #by waitingkuo
    kmeans_output = {"0_cluster":[{cancer_one:0, cancer_two:0}], "1_cluster":[{cancer_one:0, cancer_two:0}]}
    for index, row in kmeans_dataframe.iterrows():
        if row[cluster_number] == 0 and row[cancer_type] == cancer_one:
            kmeans_output["0_cluster"][0][cancer_one] +=1
        elif row[cluster_number] == 0 and row[cancer_type] == cancer_two:
            kmeans_output["0_cluster"][0][cancer_two] +=1
        elif row[cluster_number] == 1 and row[cancer_type] == cancer_one:
            kmeans_output["1_cluster"][0][cancer_one] +=1
        elif row[cluster_number] == 1 and row[cancer_type] == cancer_two:
            kmeans_output["1_cluster"][0][cancer_two] +=1          
    return kmeans_output
                       
clustering_groups = cluster_output(pheno_data, "adenocarcinoma", "squamous cell carcinoma", "characteristics_ch1.0.disease state", "ClustersfromPCA")
```

```python
clustering_groups
```

```python
def assigning_clusters(data_dictionary):
    if data_dictionary["0_cluster"][0]["adenocarcinoma"] > data_dictionary["0_cluster"][0]["squamous cell carcinoma"] and \
    data_dictionary["1_cluster"][0]["adenocarcinoma"] < data_dictionary["1_cluster"][0]["squamous cell carcinoma"]:
        adenocarcinoma_hits = data_dictionary["0_cluster"][0]["adenocarcinoma"]
        adenocarcinoma_misses = data_dictionary["0_cluster"][0]["squamous cell carcinoma"]
        squamous_hits = data_dictionary["1_cluster"][0]["squamous cell carcinoma"]
        squamous_misses = data_dictionary["1_cluster"][0]["adenocarcinoma"]
        
               
    elif data_dictionary["0_cluster"][0]["adenocarcinoma"] < data_dictionary["0_cluster"][0]["squamous cell carcinoma"] and \
    data_dictionary["1_cluster"][0]["adenocarcinoma"] > data_dictionary["1_cluster"][0]["squamous cell carcinoma"]:
        adenocarcinoma_hits = data_dictionary["1_cluster"][0]["adenocarcinoma"]
        adenocarcinoma_misses = data_dictionary["1_cluster"][0]["squamous cell carcinoma"]
        squamous_hits = data_dictionary["0_cluster"][0]["squamous cell carcinoma"]
        squamous_misses = data_dictionary["0_cluster"][0]["adenocarcinoma"]

    return [adenocarcinoma_hits, adenocarcinoma_misses, squamous_hits, squamous_misses], ["adeno_hit", "adeno_miss", "squamous_hit", "squamous_miss"]
list_of_cluster_frequencies, grouping_names = assigning_clusters(clustering_groups)
print(list_of_cluster_frequencies)
print(grouping_names)
```

```python
def computing_statistics(freq_list):
    accuracy = (freq_list[0]+freq_list[2])/sum(freq_list)
    precision_adeno = (freq_list[0]/(freq_list[0]+freq_list[1]))
    precision_squamous = (freq_list[2]/(freq_list[2]+freq_list[3]))
    recall_adeno = (freq_list[0]/(freq_list[0]+freq_list[3]))
    recall_squamous = (freq_list[2]/(freq_list[1]+freq_list[2]))
    return [accuracy, precision_adeno, precision_squamous, recall_adeno, recall_squamous], \
    ["accuracy", "precision_adeno", "precision_squamous", "recall_adeno", "recall_squamous"]

statistics, stat_names = computing_statistics(list_of_cluster_frequencies)
print(statistics, stat_names)


```
